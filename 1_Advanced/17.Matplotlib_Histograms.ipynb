```python
#importing Data Analysis Libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
import ast

#loading the dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

#data cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
df_DA_world= df[(df['job_title_short'] == 'Data Scientist')].copy()
```

```python
df_DA_world['salary_year_avg'].plot(kind='hist', bins=30, edgecolor='black' )
plt.xlim(20000, 400000)


ax = plt.gca() # Get current axis
plt.title('Distribution of Average Yearly Salary for Data Scientist Jobs in the World')
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
```
